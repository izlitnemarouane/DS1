```python
from ucimlrepo import fetch_ucirepo

# fetch dataset
heart_disease = fetch_ucirepo(id=45)

# data (as pandas dataframes)
X = heart_disease.data.features
y = heart_disease.data.targets

# metadata
print(heart_disease.metadata)

# variable information
print(heart_disease.variables)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Combine features (X) and target (y) into a single DataFrame for easier plotting
df = pd.concat([X, y], axis=1)

# Set up the matplotlib figure with a specific size for better visualization
plt.figure(figsize=(18, 12))

# --- Plot 1: Distribution of Age (Histogram) ---
# Create a subplot for the age distribution
plt.subplot(2, 3, 1)
# Use seaborn's histplot to visualize the distribution of 'age', including a Kernel Density Estimate (KDE)
sns.histplot(df['age'], kde=True)
plt.title('Distribution of Age') # Set the title of the plot
plt.xlabel('Age') # Set the x-axis label
plt.ylabel('Frequency') # Set the y-axis label

# --- Plot 2: Distribution of Sex (Countplot) ---
# Create a subplot for the sex distribution
plt.subplot(2, 3, 2)
# Use seaborn's countplot for categorical data like 'sex'
# 'hue' is set to 'sex' and 'legend=False' to avoid warnings and show counts for each category
sns.countplot(x='sex', data=df, hue='sex', palette='viridis', legend=False)
plt.title('Distribution of Sex (0=Female, 1=Male)') # Set the title with interpretation of values
plt.xlabel('Sex') # Set the x-axis label
plt.ylabel('Count') # Set the y-axis label

# --- Plot 3: Distribution of Chest Pain Type (cp) (Countplot) ---
# Create a subplot for the chest pain type distribution
plt.subplot(2, 3, 3)
# Use seaborn's countplot for categorical 'cp' (chest pain type)
sns.countplot(x='cp', data=df, hue='cp', palette='magma', legend=False)
plt.title('Distribution of Chest Pain Type') # Set the title
plt.xlabel('Chest Pain Type') # Set the x-axis label
plt.ylabel('Count') # Set the y-axis label

# --- Plot 4: Distribution of Resting Blood Pressure (trestbps) (Histogram) ---
# Create a subplot for resting blood pressure distribution
plt.subplot(2, 3, 4)
# Use histplot for 'trestbps', including KDE and a specific color
sns.histplot(df['trestbps'], kde=True, color='purple')
plt.title('Distribution of Resting Blood Pressure') # Set the title
plt.xlabel('Resting Blood Pressure (mm Hg)') # Set the x-axis label with units
plt.ylabel('Frequency') # Set the y-axis label

# --- Plot 5: Distribution of Cholesterol (chol) (Histogram) ---
# Create a subplot for cholesterol distribution
plt.subplot(2, 3, 5)
# Use histplot for 'cholesterol', including KDE and a specific color
sns.histplot(df['chol'], kde=True, color='green')
plt.title('Distribution of Cholesterol') # Set the title
plt.xlabel('Cholesterol (mg/dl)') # Set the x-axis label with units
plt.ylabel('Frequency') # Set the y-axis label

# --- Plot 6: Distribution of Heart Disease Diagnosis (num) (Countplot) ---
# Create a subplot for heart disease diagnosis distribution
plt.subplot(2, 3, 6)
# Use seaborn's countplot for the 'num' (diagnosis) target variable
sns.countplot(x='num', data=df, hue='num', palette='cividis', legend=False)
plt.title('Heart Disease Diagnosis (0=No Disease, >0=Disease)') # Set the title with interpretation
plt.xlabel('Diagnosis Value') # Set the x-axis label
plt.ylabel('Count') # Set the y-axis label

# Adjust layout to prevent plot titles/labels from overlapping
plt.tight_layout()
# Display all generated plots
plt.show()
```
